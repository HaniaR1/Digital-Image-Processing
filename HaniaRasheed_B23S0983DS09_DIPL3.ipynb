### Scenario no 01 : Traffic Sign Detection
You are working on an intelligent transportation system that detects and processes traffic signs. Your task is to

1. Convert a traffic sign image to gray scale.


- First of all we will import the necessary libraries, and load the image using cv2.imread()


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
display_image = cv2.imread('traffic light.jpeg',cv2.IMREAD_GRAYSCALE)
plt.imshow(display_image, cmap='gray')
```

2. Apply thresholding to segment the sign.

- For this we will use cv2.threshold(source image, threshold intensity value, highest value, kind of threshing.

```python
_, threshold_image=cv2.threshold(display_image,10,255,cv2.THRESH_BINARY)

plt.figure(figsize=(10,5))
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(display_image, cmap='gray')
plt.subplot(1,2,2)
plt.title('Segmented Traffic Light Image')
plt.imshow(threshold_image, cmap='gray')
plt.show()
```

3. Use edge detection to highlight the boundaries of the sign.

- For edge detection we use Canny().

```python
traffic_edges = cv2.Canny(threshold_image,10,350)
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(display_image, cmap='gray')
plt.subplot(1,2,2)
plt.title('Edges of Traffic Light Image')
plt.imshow(traffic_edges, cmap='gray')
plt.show()
```

4. Combine logical operations to remove noise and extract meaningful information.

```python
union = cv2.bitwise_or(display_image,threshold_image)
intersection = cv2.bitwise_and(display_image,255-threshold_image)
difference1 = cv2.bitwise_xor(display_image,threshold_image)


plt.figure(figsize=(24,12))
plt.subplot(1,4,1)
plt.title('Original Image')
plt.imshow(display_image, cmap='gray')
plt.subplot(1,4,2)
plt.title('union')
plt.imshow(union, cmap='gray')
plt.subplot(1,4,3)
plt.title('intersection')
plt.imshow(intersection, cmap='gray')
plt.subplot(1,4,4)
plt.title('Difference1')
plt.imshow(difference1, cmap='gray')

plt.show()
```

### Scenario no 02 : Satellite Imaging
You are working as a remote sensing analyst for an environmental research organization. Your team is studying
the impact of climate change on coastal and forested regions using satellite imagery. Your task is to process and
analyze images to extract useful insights.
1. Blend two images to create a composite image.

```python
satellite_image = cv2.imread('satellite.png',cv2.IMREAD_GRAYSCALE)
```

```python
display_height, display_width = display_image.shape

# Resize satellite_image to match display_image dimensions
satellite_image_resized = cv2.resize(satellite_image, (display_width, display_height))

# Now apply addWeighted with the resized image
alpha = 0.5  # Weight for image1
beta = 1 - alpha  # Weight for image2
blended = cv2.addWeighted(satellite_image_resized, alpha, display_image, beta, 0)
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(satellite_image, cmap='gray')
plt.subplot(1,2,2)
plt.title('Blended Satellite Image')
plt.imshow(blended, cmap='gray')
plt.show()
```

2. Apply a threshold to segment the land and water regions.  

```python
_, threshold_simage=cv2.threshold(satellite_image,140,255,cv2.THRESH_BINARY)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(satellite_image, cmap='gray')
plt.subplot(1,2,2)
plt.title('Threshold of Satellite Image')
plt.imshow(threshold_simage, cmap='gray')
plt.show()
```

3. Use edge detection to identify the boundaries of different regions.


```python
satellite_edges = cv2.Canny(satellite_image,50,150)
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(satellite_image, cmap='gray')
plt.subplot(1,2,2)
plt.title('Edges of Satellite Image')
plt.imshow(satellite_edges, cmap='gray')
plt.show()
```

4. Create a mask to highlight a specific region of interest (e.g., a forest area).

```python
lower_green = 20
upper_green = 90

# Create a mask for green regions
mask = cv2.inRange(satellite_image, lower_green, upper_green)

# Apply the mask to the original image to highlight green regions
highlighted = cv2.bitwise_and(satellite_image, satellite_image, mask=mask)


plt.figure(figsize=(15, 10))
plt.subplot(1, 3, 1)
plt.imshow(satellite_image, cmap='gray')
plt.title("Original Image")

plt.subplot(1, 3, 2)
plt.imshow(mask, cmap="gray")
plt.title("Mask for Green Areas")

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(highlighted, cv2.COLOR_BGR2RGB))
plt.title("Highlighted Green Areas")


plt.show()
```
