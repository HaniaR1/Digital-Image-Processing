### Task 4 Crop the Image
- Crop a specific section of the image (coordinates covering a clear patch of open ground) and save it as
open_patch.jpg.

#### Description:
For this we first read the image as cv2, then we crop it by specifying the values of x1, y1, x2 and y2. Then we write the updated cropped image using imwrite.

```python
image=cv2.imread('forest.jpeg')
x1,y1,x2,y2 = 80, 10, 160, 60
open_patch= image[y1:y2,x1:x2]
cv2.imwrite('open_patch.jpg',open_patch)
print("cropped section saved")

cv2_imshow(image)
print("Cropped image is:")
cv2_imshow(open_patch)
```

### Task 6 Draw Bounding Boxes:
-  Highlight three clear patches in the original image by drawing rectangles around them. Save the result as
highlighted_forest_area.jpg.

#### Description:
Cv2 reads the file in a BRG fashion, we create the bouding boxes by specifying indiviudal x1,y1,x2 and y2 values. Once the boxes are created we convert the image back to RGB for better display. After this we save and show the image

```python
image = cv2.imread('forest.jpeg')
x1, y1, x2, y2 = 80, 10, 150, 60
cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 10)

x1, y1, x2, y2 = 120, 100, 150, 180
cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 10)

x1, y1, x2, y2 = 0, 40, 20, 120
cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)

# Convert the image to RGB for proper display
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6, 6))
plt.imshow(image_rgb)
plt.axis('off')
plt.title("Highlighted Forest Area")

cv2.imwrite('highlighted_forest_area.jpg', image)
plt.show()
```
